- 샤이썬(서울대 수강신청 빈자리 알리미) 같은 기능을 만들고자 함 

```python
import requests
import json

url = 'https://kauth.kakao.com/oauth/token'
client_id = ''
redirect_uri = ''
code = ''
data = {
    'grant_type':'authorization_code',
    'client_id':client_id,
    'redirect_uri':redirect_uri,
    'code': code,
    }

response = requests.post(url, data=data)
tokens = response.json()

#발행된 토큰 저장
with open("token.json","w") as kakao:
    json.dump(tokens, kakao)
```

```python
def send_kakao_text(room_text):
        with open("token.json","r") as kakao:
            tokens = json.load(kakao)

        url="https://kapi.kakao.com/v2/api/talk/memo/default/send"

        headers={
            "Authorization" : "Bearer " + tokens["access_token"]
        }

        data = {
               'object_type': 'text',
               'text': room_text,
               'link': {
                   'web_url': 'https://developers.kakao.com',
                   'mobile_web_url': 'https://developers.kakao.com'
               },
           }

        data = {'template_object': json.dumps(data)}
        response = requests.post(url, headers=headers, data=data)
        response.status_code
```

## 1. 키이스케이프

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome('chromedriver.exe')
url = 'https://keyescape.co.kr/web/home.php?go=rev.make'
driver.get(url)
time.sleep(3)
repeat = 0

while True:
    for spot in driver.find_elements_by_css_selector('div div > div > div > div > dl > dd > div > ul > a')[:2]: # 우주라이크, 더오름
        repeat += 1
        # 하나씩 클릭
        spot_text = spot.text
        spot.send_keys(Keys.ENTER)
        time.sleep(3)
        # 날짜 클릭
        for dat1 in driver.find_elements_by_css_selector('div div > div > div > div > dl#rev2.rev2 table.t_calendar td.day_index > a'):
            date_text = dat1.text
            dat1.send_keys(Keys.ENTER)
            time.sleep(3)
            # 테마 클릭
            for theme in driver.find_elements_by_css_selector('div div > div > div > div > dl.rev3 ul#theme_data.revlist > a'):
                theme_text = theme.text
                theme.send_keys(Keys.ENTER)
                time.sleep(3)
                # 시간 비는지 확인
                for postime in driver.find_elements_by_css_selector('div div > div > div > div > dl.rev4 ul#theme_time_data.revlist > a'):
                    time_text = postime.text
                    room_data = "위치:"+spot_text+" / 날짜:"+date_text+" / 테마:"+theme_text+' / 시간:'+time_text
                    send_kakao_text(room_data)
            
            
        ## 다음 월로 이동
        if repeat%2 == 1:
            next_button = driver.find_element_by_css_selector('div div > div > div > div > dl#rev2.rev2 div.calendar_yymm > li.next > a')
            next_button.send_keys(Keys.ENTER)
            time.sleep(3)
            # 날짜 클릭
            for dat1 in driver.find_elements_by_css_selector('div div > div > div > div > dl#rev2.rev2 table.t_calendar td.day_index > a'):
                date_text = dat1.text
                dat1.send_keys(Keys.ENTER)
                time.sleep(3)
                # 테마 클릭
                for theme in driver.find_elements_by_css_selector('div div > div > div > div > dl.rev3 ul#theme_data.revlist > a'):
                    theme_text = theme.text
                    theme.send_keys(Keys.ENTER)
                    time.sleep(3)
                    # 시간 비는지 확인
                    for postime in driver.find_elements_by_css_selector('div div > div > div > div > dl.rev4 ul#theme_time_data.revlist > a'):
                        time_text = postime.text
                        room_data = "위치:"+spot_text+" / 날짜:"+date_text+" / 테마:"+theme_text+' / 시간:'+time_text
                        send_kakao_text(room_data)
                    
        ## 이전 월로 이동
        else:
            prev_button = driver.find_element_by_css_selector('div div > div > div > div > dl#rev2.rev2 div.calendar_yymm > li.prev > a')
            prev_button.send_keys(Keys.ENTER)
            time.sleep(3)
            # 날짜 클릭
            for dat1 in driver.find_elements_by_css_selector('div div > div > div > div > dl#rev2.rev2 table.t_calendar td.day_index > a'):
                date_text = dat1.text
                dat1.send_keys(Keys.ENTER)
                time.sleep(3)
                # 테마 클릭
                for theme in driver.find_elements_by_css_selector('div div > div > div > div > dl.rev3 ul#theme_data.revlist > a'):
                    theme_text = theme.text
                    theme.send_keys(Keys.ENTER)
                    time.sleep(3)
                    # 시간 비는지 확인
                    for postime in driver.find_elements_by_css_selector('div div > div > div > div > dl.rev4 ul#theme_time_data.revlist > a'):
                        time_text = postime.text
                        room_data = "위치:"+spot_text+" / 날짜:"+date_text+" / 테마:"+theme_text+' / 시간:'+time_text
                        send_kakao_text(room_data)
    
    # 10초마다 새로고침 반복
    time.sleep(10)
    driver.refresh()
```
